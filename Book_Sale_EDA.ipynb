## 1 Introduction

Le **dataset des livres : ventes, évaluations et publication** fournit des informations complètes sur divers aspects des livres.  
Ce dataset permet d’analyser le marché du livre, d’évaluer la performance des auteurs et de découvrir des tendances dans les préférences de lecture.

---

### 1.1 Tableau des colonnes

| Colonne (anglais)        | Colonne (français)       | Description |
|--------------------------|-------------------------|-------------|
| Publishing Year          | Année de publication     | L’année où chaque livre a été publié. Permet d’analyser la répartition chronologique. |
| Book Name                | Nom du livre            | Le titre du livre, unique pour chaque ouvrage. |
| Author                   | Auteur                  | Le(s) nom(s) de l’auteur ou des auteurs. Utile pour analyser leurs contributions et leur réputation. |
| language_code            | Code langue             | Code représentant la langue du livre. Permet des analyses basées sur la langue. |
| Author_Rating            | Note de l’auteur        | Note attribuée à l’auteur basée sur ses œuvres précédentes. |
| Book_average_rating      | Note moyenne du livre   | Moyenne des notes données par les lecteurs pour chaque livre. |
| Book_ratings_count       | Nombre d’évaluations    | Nombre d’évaluations reçues par chaque livre, mesure l’engagement des lecteurs. |
| genre                    | Genre                   | Catégorie littéraire ou genre du livre. Permet d’analyser les tendances par genre. |
| gross sales              | Ventes brutes           | Chiffre d’affaires total généré par le livre. |
| publisher revenue        | Revenus de l’éditeur    | Revenus que l’éditeur tire de la vente de chaque livre. |
| sale price               | Prix de vente           | Prix auquel chaque exemplaire du livre est vendu. |
| sales rank               | Classement des ventes   | Rang du livre selon ses ventes comparé aux autres livres du même genre. |
| units sold               | Unités vendues          | Nombre total d’exemplaires vendus pour chaque titre. |

## 4. Nettoyage des donnees

```python
## Verifier s'il y'a une valeur manquante 
df.isnull().any()
```

```python
## visualiser les lignes ayant des valeurs manquantes (Publisher Year , Book name , language code)
df[df[['Publishing Year', 'Book Name', 'language_code']].isnull().any(axis=1)].head(10)
```

```python
## le nombre de valeur manquante pas colonnes
df[['Publishing Year', 'Book Name', 'language_code']].isnull().sum()
```

```python
## Supprimer la ligne où Publishing Year est manquant
df = df.dropna(subset=['Publishing Year'])

## Remplacer les valeurs manquantes de Book Name par "Unknown"
df['Book Name'] = df['Book Name'].fillna('Unknown')

## Remplacer les valeurs manquantes de language_code par "Unknown"
df['language_code'] = df['language_code'].fillna('Unknown')

# Vérification du résultat
df[['Publishing Year', 'Book Name', 'language_code']].isnull().sum()
```

```python
## convertir Publishing Year en entier 
df['Publishing Year'] = df['Publishing Year'].astype('Int64')  # gère les valeurs manquantes
df['Publishing Year'].dtypes
```

## 5. Visualisation


```python
## style global des graphiques
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

## creation d'un fichier 
import os
output_dir = "output"
os.makedirs(output_dir, exist_ok=True)
```

```python
## 5.1 Distribution des notes moyennes des livres
plt.figure()

sns.histplot(df["Book_average_rating"], bins=20, kde=True)

plt.title("Distribution des notes moyennes des livres")
plt.xlabel("Note moyenne")
plt.ylabel("Nombre de livres")

plt.savefig(f"{output_dir}/distribution_notes_livres.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# 5.2 Top 10 auteurs par unités vendues
top_authors = (
    df.groupby("Author")["units sold"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure()

sns.barplot(x=top_authors.values, y=top_authors.index)

plt.title("Top 10 des auteurs par unités vendues")
plt.xlabel("Unités vendues")
plt.ylabel("Auteur")

plt.savefig(f"{output_dir}/top_auteurs_unites_vendues.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
## Revenus par genre
genre_sales = df.groupby("genre")["gross sales"].sum().sort_values(ascending=False)

plt.figure()

sns.barplot(x=genre_sales.values, y=genre_sales.index)

plt.title("Revenus générés par genre")
plt.xlabel("Ventes brutes")
plt.ylabel("Genre")

plt.savefig(f"{output_dir}/revenus_par_genre.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# 7. Relation prix vs unités vendues
plt.figure()

sns.scatterplot(
    data=df,
    x="sale price",
    y="units sold"
)

plt.title("Relation entre le prix du livre et les unités vendues")
plt.xlabel("Prix de vente")
plt.ylabel("Unités vendues")

plt.savefig(f"{output_dir}/prix_vs_unites_vendues.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# 8. Notes moyennes par genre
plt.figure()

sns.boxplot(
    data=df,
    x="genre",
    y="Book_average_rating"
)

plt.title("Distribution des notes moyennes par genre")
plt.xlabel("Genre")
plt.ylabel("Note moyenne")

plt.xticks(rotation=45)

plt.savefig(f"{output_dir}/notes_par_genre.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# 9. Livres publiés par année
plt.figure()

sns.countplot(
    data=df,
    x="Publishing Year"
)

plt.title("Nombre de livres publiés par année")
plt.xlabel("Année de publication")
plt.ylabel("Nombre de livres")

plt.xticks(rotation=45)

plt.savefig(f"{output_dir}/livres_par_annee.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# 10. Note auteur vs note livre
plt.figure()

sns.scatterplot(
    data=df,
    x="Author_Rating",
    y="Book_average_rating"
)

plt.title("Relation entre la note de l'auteur et la note moyenne du livre")
plt.xlabel("Note de l'auteur")
plt.ylabel("Note moyenne du livre")

plt.savefig(f"{output_dir}/note_auteur_vs_note_livre.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
#11. Nombre d'avis vs note moyenne
plt.figure()

sns.scatterplot(
    data=df,
    x="Book_ratings_count",
    y="Book_average_rating"
)

plt.title("Relation entre le nombre d'avis et la note moyenne")
plt.xlabel("Nombre d'avis")
plt.ylabel("Note moyenne")

plt.savefig(f"{output_dir}/nombre_avis_vs_note.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# Dataset contenant uniquement les variables numériques
df_numeric = df.select_dtypes(include=["int64", "float64"])

corr_matrix = df_numeric.corr()

plt.figure(figsize=(12,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Heatmap de corrélation des variables numériques")

plt.savefig(f"{output_dir}/heatmap_correlation.png", dpi=300, bbox_inches="tight")

plt.show()
```
