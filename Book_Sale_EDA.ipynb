## 1 Introduction

Le **dataset des livres : ventes, évaluations et publication** fournit des informations complètes sur divers aspects des livres.  
Ce dataset permet d’analyser le marché du livre, d’évaluer la performance des auteurs et de découvrir des tendances dans les préférences de lecture.

---

### 1.1 Tableau des colonnes

| Colonne (anglais)        | Colonne (français)       | Description |
|--------------------------|-------------------------|-------------|
| Publishing Year          | Année de publication     | L’année où chaque livre a été publié. Permet d’analyser la répartition chronologique. |
| Book Name                | Nom du livre            | Le titre du livre, unique pour chaque ouvrage. |
| Author                   | Auteur                  | Le(s) nom(s) de l’auteur ou des auteurs. Utile pour analyser leurs contributions et leur réputation. |
| language_code            | Code langue             | Code représentant la langue du livre. Permet des analyses basées sur la langue. |
| Author_Rating            | Note de l’auteur        | Note attribuée à l’auteur basée sur ses œuvres précédentes. |
| Book_average_rating      | Note moyenne du livre   | Moyenne des notes données par les lecteurs pour chaque livre. |
| Book_ratings_count       | Nombre d’évaluations    | Nombre d’évaluations reçues par chaque livre, mesure l’engagement des lecteurs. |
| genre                    | Genre                   | Catégorie littéraire ou genre du livre. Permet d’analyser les tendances par genre. |
| gross sales              | Ventes brutes           | Chiffre d’affaires total généré par le livre. |
| publisher revenue        | Revenus de l’éditeur    | Revenus que l’éditeur tire de la vente de chaque livre. |
| sale price               | Prix de vente           | Prix auquel chaque exemplaire du livre est vendu. |
| sales rank               | Classement des ventes   | Rang du livre selon ses ventes comparé aux autres livres du même genre. |
| units sold               | Unités vendues          | Nombre total d’exemplaires vendus pour chaque titre. |

## 4. Nettoyage des donnees

```python
## Verifier s'il y'a une valeur manquante 
df.isnull().any()
```

```python
## visualiser les lignes ayant des valeurs manquantes (Publisher Year , Book name , language code)
df[df[['Publishing Year', 'Book Name', 'language_code']].isnull().any(axis=1)].head(10)
```

```python
## le nombre de valeur manquante pas colonnes
df[['Publishing Year', 'Book Name', 'language_code']].isnull().sum()
```

```python
## Supprimer la ligne où Publishing Year est manquant
df = df.dropna(subset=['Publishing Year'])

## Remplacer les valeurs manquantes de Book Name par "Unknown"
df['Book Name'] = df['Book Name'].fillna('Unknown')

## Remplacer les valeurs manquantes de language_code par "Unknown"
df['language_code'] = df['language_code'].fillna('Unknown')

# Vérification du résultat
df[['Publishing Year', 'Book Name', 'language_code']].isnull().sum()
```

```python
## convertir Publishing Year en entier 
df['Publishing Year'] = df['Publishing Year'].astype('Int64')  # gère les valeurs manquantes
df['Publishing Year'].dtypes
```
